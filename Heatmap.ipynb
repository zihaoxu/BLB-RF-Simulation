```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
from collections import defaultdict
sns.set_style("ticks")
```

```python
params = {'axes.labelsize': 'large',
          'axes.titlesize':'large'}       ###'axes.titlesize':'xx-large'
pylab.rcParams.update(params)
```

```python
df = pd.read_csv('/Users/zihaoxu/R_repos/BLB-RF-Sim/CSV/heatmap_clustered_0.50.6.csv')
if 'Unnamed: 0' in list(df):
        df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df
```

```python
df5 = df[df.gamma==0.6]
```

```python
for ga in [.5,.6]:
    MSE_data = df5.pivot(index = 'ntree', columns = 's', values = 'MSE')
    sns.heatmap(MSE_data,annot = True)
    plt.title('MSE(gamma = ' + str(ga) + ')')
    plt.savefig('/Users/zihaoxu/R_repos/BLB-RF-Sim/Plots/MSE~ntreeXs'+str(ga)+'.png', dpi = 300, fmt = 'png')
    plt.show()
```

```python
for ga in [.5,.6]:
    time_data = df5.pivot(index = 'ntree', columns = 's', values = 'time')
    sns.heatmap(time_data,annot = True)
        plt.title('Time(gamma = ' + str(ga) + ')')
    plt.savefig('/Users/zihaoxu/R_repos/BLB-RF-Sim/Plots/MSE~ntreeXs'+str(ga)+'.png', dpi = 300, fmt = 'png')
    plt.show()
```
