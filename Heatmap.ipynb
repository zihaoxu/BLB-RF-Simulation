```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
from collections import defaultdict
sns.set_style("ticks")
```

```python
params = {'axes.labelsize': 'x-large',
          'axes.titlesize':'x-large'}       ###'axes.titlesize':'xx-large'
pylab.rcParams.update(params)
```

```python
df = pd.read_csv('/Users/zihaoxu/R_repos/BLB-RF-Sim/CSV/heatmap_cosine.csv')
if 'Unnamed: 0' in list(df):
        df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
df.head()
```

```python
for ga in df.gamma.unique():
    df5 = df[df.gamma==ga]
    MSE_data = df5.pivot(index = 'ntree', columns = 's', values = 'MSE')
    sns.heatmap(MSE_data,annot = None)
    plt.title('MSE(gamma = ' + str(ga) + ')')
    #plt.savefig('/Users/zihaoxu/R_repos/BLB-RF-Sim/Plots/MSE~ntreeXs'+str(ga)+'.png', dpi = 300, fmt = 'png')
    plt.show()
```

```python
for ga in df.gamma.unique():
    df5 = df[df.gamma==ga]
    time_data = df5.pivot(index = 'ntree', columns = 's', values = 'time')
    sns.heatmap(time_data,annot = True)
    plt.title('Time (gamma = ' + str(ga) + ')')
    plt.savefig('/Users/zihaoxu/R_repos/BLB-RF-Sim/Plots/Time~ntree+s'+str(ga)+'.png', dpi = 300, fmt = 'png')
    plt.show()
```
