```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/zihaoxu/Desktop/result.csv')
```

```python
df.head()
```

```python
plt.scatter(x = df['yexp'], y = df['yhat'], c = 'b', label = 'BLB-RF')
plt.scatter(x = df['real_y_rf'], y = df['rf_result'], c = 'g', label = 'RF')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.plot([495,545], [495,545], lw = 1, c = 'grey')
plt.xlim(495,545)
plt.ylim(495,545)
plt.gca().set_aspect('equal')
plt.title('Performance of BLB-RF vs RF')
plt.legend(loc=2)


sns.despine()

#plt.savefig('/Users/zihaoxu/Desktop/Performance_linear.png', dpi = 300, fmt = 'png')
plt.show()
```

```python
sns.distplot(df['diff'], label = 'BLB-RF')
sns.distplot(df['real_y_rf'] - df['rf_result'], label = 'RF')
plt.legend()
plt.title('Residual Distribution')
#plt.savefig('/Users/zihaoxu/Desktop/Residuals_linear.png', dpi = 300, fmt = 'png')
plt.show()

```
